```python
import dash
import dash_core_components as dcc
import dash_html_components as html
import plotly.graph_objs as go

mgr_options = population['year'].unique()

app = dash.Dash()

app.layout = html.Div([
    html.H2("Comparision between Dallas and Houston"),
    html.Div(
        [
            dcc.Dropdown(
                id="year",
                options=[{
                    'label': i,
                    'value': i
                } for i in mgr_options],
                value='All years'),
        ],
        style={'width': '25%',
               'display': 'inline-block'}),
    dcc.Graph(id='funnel-graph'),
])
```

```python
@app.callback(
    dash.dependencies.Output('funnel-graph', 'figure'),
    [dash.dependencies.Input('year', 'value')])
def update_graph(year):
    if year == "All years":
        df_plot = population[population['year']==2016]
    else:
        df_plot = population[population['year'] == year]

    pv = pd.pivot_table(
        df_plot,
        index=['city'],
        columns=['year'],
        values=['male'],
        aggfunc=sum,
        fill_value=0)
    trace1 = go.Bar(x=pv.index, y=pv[('male', 2016)], name='2016')
    trace2 = go.Bar(x=pv.index, y=pv[('male', 2017)], name='2017')
    trace3 = go.Bar(x=pv.index, y=pv[('male', 2018)], name='2018')
    trace4 = go.Bar(x=pv.index, y=pv[('male', 2019)], name='2019')

    return {
        'data': [trace1, trace2, trace3, trace4],
        'layout':
        go.Layout(
            title='Comparision between Houston and Dallas in the year {}'.format(year),
            barmode='stack')
    }


if __name__ == '__main__':
    app.run_server(host='127.0.0.1',port=8080,debug=True)
```
