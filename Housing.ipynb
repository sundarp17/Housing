```python
states = states.drop(columns='created')
states.head()
```

```python
states = states.rename(columns={"states_id":"state_id"})
states.head()
```

```python
#merging state and state_census data
state_data = pd.merge(states,state_census,on='state_id')
state_data.head()
```

```python
#selecting one particular state
texas = state_data[state_data['state_id']==44]

texas.reset_index(inplace=True)
texas
```

```python
df_texas = texas[['white_alone','black_africal','native','asian','islander','other','year']]
df_texas = df_texas.set_index('year')
df_texas
```

```python
#visualizing the types of people in texas from 2015-2019.

from matplotlib import pyplot as plt
plt.rcParams["figure.figsize"] = [14, 4]
df_texas.plot(kind='bar')
plt.show()
```
